```python
import sys
sys.path.append('../src/')

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from DataSetFactory import DataSetFactory
from ClusterPlot import ClusterPlot
```

```python
RANDOM_STATE = 42
```

```python
ds = DataSetFactory.get_dataset('FashionMNIST', random_state=RANDOM_STATE, is_subset=[0,2,8], sample=0.15)
ds_64 = DataSetFactory.get_dataset('FashionMNIST64', random_state=RANDOM_STATE, is_subset=[0,2,8], sample=0.15)
```

```python
outliers = [15,25,35]
```

```python
for i in outliers:
    print(i)
    im = ds.df[ds.feature_cols].iloc[i].values.reshape(28, 28)
    plt.imshow(im, cmap='gray')
    plt.show()
    file_name = f'./plots/FashionMNISTAnnotation/image_{i}.png'
    plt.imsave(file_name, im, cmap='gray')
    
    im = ds_64.df[ds_64.feature_cols].iloc[i].values.reshape(8, 8)
    plt.imshow(im, cmap='gray')
    plt.show()
```
